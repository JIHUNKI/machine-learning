```python
import pandas as pd
import numpy as np


df= pd.read_csv('./datasets/sonar.csv',header=None)
df.head()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
```

```python
dataset = df.values

x = dataset[:, 0:60]
x = np.asarray(x).astype(np.float32) # dtype이 object인데 float으로 변경
y_obj = dataset[:, 60]
# 문자열 변환
e = LabelEncoder()
e.fit(y_obj)
y = e.transform(y_obj)
# 모델 설정
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# 모델 컴파일
model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])

# 모델 실행
# model.fit(x, y, epochs=200, batch_size=5)
# # 결과 출력
# print('\n Accuracy: %.4f' % (model.evaluate(x,y)[1]))
```

```python
history = model.fit(x, y, epochs=200, batch_size=5,validation_split=0.3)
```

```python
result = history.history
```

```python
## matplotlib을 이용해서 그래프
import matplotlib.pyplot as plt

fig,ax = plt.subplots()
ax.plot(result['val_loss'][0:25],label ='val_loss')
ax.plot(result['accuracy'][0:25],label='accuracy')
plt.legend()
ax.set_xlabel('epochs')
ax.set_title('validation ')

# epoch를 20정도가 과적합을 막을수 있음
```

```python
dataset = df.values

x = dataset[:, 0:60]
x = np.asarray(x).astype(np.float32) # dtype이 object인데 float으로 변경
y_obj = dataset[:, 60]
# 문자열 변환
e = LabelEncoder()
e.fit(y_obj)
y = e.transform(y_obj)

## 학습셋 과 테스트 셋의 구분
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=120)
# 모델 설정
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# 모델 컴파일
model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])

# 모델 실행
model.fit(x_train, y_train, epochs=130, batch_size=5)
model.save('model.h5')
```

```python
## 저장된 모델 로드
from tensorflow.keras.models import load_model
model = load_model('./model.h5')
```

```python
model.evaluate(x_test,y_test)
```
