```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as ts
import numpy as np
import matplotlib.pyplot as plt

# 가상데이터 생성
x = np.linspace(1,2,200)
y = x*4+np.random.randn(200)*0.3 ## 약간의 편차를 줌

model = Sequential()
model.add(Dense(1,input_dim=1,activation='linear'))

model.compile(optimizer='sgd',loss='mse',metrics=['mse'])
model.fit(x,y,batch_size=1,epochs=30)

predict =model.predict(x)

plt.plot(x,predict,'blue',x,y,'k.') # 첫번째 그래프는 파란색 마커로
plt.show()                       # 두번째 그래프는 검정색.으로 그린다
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = [[2,81],[4,93],[6,91],[8,97]]
x= [i[0] for i in data]
y = [i[1] for i in data]

plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.show()
```

```python
x_data = np.array(x)
y_data = np.array(y)

# 초기 회귀계수
a =0
b= 0

# 학습률 
lr =0.03

# 몇번 반복할지
epochs = 2023

# 경사하강법
for i in range(epochs):
    y_hat = a*x_data+b
    error = y_data-y_hat
    a_diff = -(2/len(x_data)) * sum(x_data *(error))
    b_diff = -(2/len(x_data)) * sum((error))
    a = a - lr * a_diff
    b = b - lr * b_diff
    
    ## 변화된 회귀계수에 대한 error
    if i % 100 ==0:
        mse = np.mean((error)**2)
        print(f'epoch = {i}, 기울기 = {np.round(a,4)}, 절편 = {np.round(b,4)}, mse ={np.round(mse,4)}')
```

```python
y_pred = a *x_data +b
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.plot(x_data,y_pred)
plt.show()
```

```python
data = [[2,0,81],[4,4,93],[6,2,91],[8,3,97]]
x1 = [i[0] for i in data]
x2 = [i[1] for i in data]
y = [i[2] for i in data]

x1_data = np.array(x1)
x2_data = np.array(x2)
y_data = np.array(y)

# 초기 회귀계수
a1 =0
a2 =0
b= 0

# 학습률 
lr =0.02

# 몇번 반복할지
epochs = 2023

# 경사하강법
for i in range(epochs):
    y_hat = a1*x1_data+a2*x2_data +b
    error = y_data-y_hat
    a1_diff = -(2/len(x1_data)) * sum(x1_data *(error))
    a2_diff = -(2/len(x2_data)) * sum(x2_data *(error))
    b_diff = -(2/len(x1_data)) * sum((error))
    a1 = a1 - lr * a1_diff
    a2 = a2 - lr * a2_diff
    b = b - lr * b_diff
    
    ## 변화된 회귀계수에 대한 error
    if i % 100 ==0:
        mse = np.mean((error)**2)
        print(f'epoch = {i}, a1 = {np.round(a1,4)},a2 = {np.round(a2,4)}, 절편 = {np.round(b,4)}, mse ={np.round(mse,4)}')
```
