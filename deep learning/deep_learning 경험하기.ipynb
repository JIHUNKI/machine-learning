```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow import one_hot

import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()

x = iris.data
y = iris.target
one_hot_y = one_hot(y,depth=3)


# 딥러닝 구조를 결정합니다(모델을 설정하고 실행하는 부분)
model = Sequential()
## 입력 변수 4개
model.add(Dense(30,input_dim=4,activation='relu'))
# iris는 y 출력물이 3개 이므로 마지막 노드는 3개 
# 다중 클래스 이므로활성함수로 소프트 맥스 사용
model.add(Dense(3,activation='softmax'))

# 딥러닝 실행
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(x,one_hot_y,epochs=100,batch_size=10)
 # 반복 10번 한번할때 10개행씩 계산
```
